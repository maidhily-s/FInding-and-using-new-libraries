```python
import matplotlib.pyplot as plt
```

```python
fruits = ['Apple', 'Banana', 'Mango', 'Orange', 'Grapes']
ratings = [8, 6, 10, 7, 9]
```

```python
plt.bar(fruits, ratings, color='skyblue')
plt.title("My Favorite Fruits")
plt.xlabel("Fruits")
plt.ylabel("How Much I Like Them (out of 10)")
```
